## Plotly and Cufflinks

Plotly is a library that allows you to create interactive plots that you can use in dashboards or websites (you can save them as html files or static images).

Installation
In order for this all to work, you will need to install plotly and cufflinks to call plots directly off of a pandas dataframe. These libraries are not currently available through conda but are available through pip. Install the libraries at your command line/terminal using:

pip install plotly 
pip install cufflinks
** NOTE: Make sure you only have one installation of Python on your computer when you do this, otherwise the installation may not work. **

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
```

```python
from plotly import __version__
```

```python
print(__version__)
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

```python
df.sum().iplot(kind='bar')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.scatter_matrix()
```
